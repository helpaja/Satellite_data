```python
#import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import seaborn as sns
import scipy
#import imageio
#import imageio.v3 as iio
from pathlib import Path
import os
from scipy.optimize import curve_fit
import pandas
import xarray
import cartopy.crs as ccrs
from netCDF4 import Dataset
import cartopy.feature as cfeature
```

```python

file = "S5P_PAL__L2__SIF____20251231T231651_20260101T005821_42585_03_010001_20260102T172631.nc"

# Avaa PRODUCT-ryhmä
ds = xarray.open_dataset(file, group="PRODUCT")
```

```python
lat2 = ds["latitude"]
lon2 = ds["longitude"]
sif2 = ds["SIF_743"]
```

```python
fig, axs = plt.subplots(1,1,figsize=(6,6), subplot_kw={'projection': ccrs.PlateCarree()})
ax = axs
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.add_feature(cfeature.LAND, alpha=0.5)
ax.add_feature(cfeature.OCEAN, facecolor='lightblue', alpha = 0.5)
ax.gridlines(draw_labels=True)
#ax.set_extent([15, 35, 58, 72],crs=ccrs.PlateCarree())
scatter = plt.scatter(lon2, lat2, c=sif2, s=5)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("SIF_743 Map")
plt.colorbar(scatter, label="SIF_743")

plt.show()
```
